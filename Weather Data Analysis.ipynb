### Exploratory Data Analysis

Find the unique wind speed values in data.

```python
data['Wind Speed_km/h'].unique()
```

Find no of times when weather is exactly clear.

```python
data.Weather.value_counts()
```

```python
data[data.Weather == 'Clear']
```

```python
data.groupby('Weather').get_group('Clear')
```

Find no of times when wind speed was exactly 4 km/hr.

```python
data[data['Wind Speed_km/h'] == 4]
```

Find out null values in data.

```python
data.isnull().sum()
```

```python
data.notnull().sum()
```

Rename column weather of dataframe to weather condition.

```python
data.rename(columns = {'Weather' : 'Weather Condition'}, inplace = True)
data.head()
```

What is the mean visibility?

```python
data.Visibility_km.mean()
```

What is standard deviation of pressure in this data?

```python
data.Press_kPa.std()
```

What is variance of relative humidity in this data?

```python
data['Rel Hum_%'].var()
```

Find all instances when snow was recorded.

```python
data['Weather Condition'].value_counts()
```

```python
data[data['Weather Condition'] == 'Snow']
```

```python
data[data['Weather Condition'].str.contains('Snow')]
```

Find out all instances when wind speed is above 24 & visibility is 25.

```python
data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)]
```

What is mean value for each column against each weather condition?

```python
data.groupby('Weather Condition').mean()
```

What is minimum & maximum value of each column against each weather condition?

```python
data.groupby('Weather Condition').min()
```

```python
data.groupby('Weather Condition').max()
```

Show all records where weather condition is fog.

```python
data[data['Weather Condition'] == 'Fog']
```

Find all instances when weather is clear or visibility is above 40.

```python
data[(data['Weather Condition'] == 'Clear') | (data['Visibility_km'] > 40)]
```

Find all instances when:
1) weather is clear & relative humidity is greater than 50

or

2) visibility is above 40

```python
data[(data['Weather Condition'] == 'Clear') & (data['Rel Hum_%'] > 50) | (data['Visibility_km'] > 40)]
```
