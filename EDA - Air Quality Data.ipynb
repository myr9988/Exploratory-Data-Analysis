### Importing necessary Libraries & Dataset

```python
import pandas as pd

data = pd.read_csv(r"C:\Users\SHREE\Downloads\Python CODES\EDA - Air Quality Data\air-quality-data.csv")
data.head()
```

```python
data.info()
```

### Data Preprocessing

#### Converting object data type column to date time data type

```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

```python
data.info()
```

### Show the Year-wise average pollution with Line Chart

```python
data['Year'] = data.Timestamp.dt.year
data
```

```python
year_avg = data.groupby('Year')['PM2.5'].mean()
year_avg
```

#### Year-wise average pollution Line Chart

```python
year_avg.plot(kind = 'line', figsize = (15,5));
```

```python
year_avg.plot(kind = 'line', figsize = (15,5), style = '*');
```

### Draw the Area plot showing average pollution Month wise

```python
data['Month'] = data.Timestamp.dt.month
data
```

```python
month_avg = data.groupby('Month')['PM2.5'].mean()
month_avg
```

#### Area plot showing average pollution Month wise

```python
month_avg.plot(kind = 'area', figsize = (12,5), color = 'orange');
```

### Draw a Bar graph to show the Hour-wise average pollution

```python
data['Hour'] = data.Timestamp.dt.hour
data
```

```python
hour_avg = data.groupby('Hour')['PM2.5'].mean()
hour_avg
```

#### Bar graph to show the Hour-wise average pollution

```python
hour_avg.plot(kind = 'bar', figsize = (15,5), color = 'purple');
```

### Usually in which month/s the air was recorded 'Very Unhealthy'?

```python
air_very_unhealthy = data[(data['PM2.5'] >= 150.5) & (data['PM2.5'] <= 250.4)]
air_very_unhealthy
```

```python
air_very_unhealthy.value_counts('Month')
```

### Usually in which month/s the air was fresh(Good)?

```python
air_good = data[data['PM2.5'] <= 12.0]
air_good
```

```python
air_good.value_counts('Month')
```

### In year 2018, how many times AQI wa recorded 'Moderate'?

```python
air_moderate = data[(data['Year'] == 2018) & (data['PM2.5'] >= 12.1) & (data['PM2.5'] <= 35.4)]
air_moderate
```

```python
air_moderate.Year.unique()
```
